- Here we can see that both length and width of sepal and petals are in numeric form but species is in string

- Machine learning algorithms typically require numeric data as input. Since the 'species' column in the IRIS dataset contains categorical string labels ('setosa', 'versicolor', 'virginica'), we need to convert it into numerical form before training the model.

One common technique to convert categorical data into numerical form is called "label encoding.

# Label Encoder

- Converts the labels into numeric form so as to convert it into the machine-readable form
- Label encoding assigns a unique integer to each category in the column. In this case, 'setosa' might be encoded as 0, 'versicolor' as 1, and 'virginica' as 2.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# Use Label Encoder to convert 'species' column to numerical form
iris_df['species'] = le.fit_transform(iris_df['species'])

```

```python
iris_df.sample(5)
```

#### Model Training

```python
from sklearn.model_selection import train_test_split
# Separate features (attributes) and target (class) from the DataFrame
X = iris_df.drop(columns=['species'])
y = iris_df['species']
# Separate the data into train - 80% and test - 20%
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20)

```

#### Using Logistic Regression

```python
# Create and train the Logistic Regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
# Make predictions on the test sets
predicted = model.predict(X_test)
```

- The result is converted back to the original categorical label using the 'inverse_transform()' method of the LabelEncoder.

```python
predicted_label = le.inverse_transform(predicted)
print(predicted_label)
```

#### Evaluate the model

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Accuracy
accuracy = accuracy_score(y_test,predicted)
print("Accuracy",accuracy)
```

```python
# Classification Report
classification_rep = classification_report(y_test,predicted)
print("Classification Report:\n", classification_rep) 
```

- For class 0 ('setosa'), precision, recall, and F1-score are all 1.0, indicating that the model perfectly predicts this class.
- For class 1 ('versicolor'), precision is 1.0, meaning all positive predictions are correct, recall is 0.93, meaning it correctly identifies 93% of actual 'versicolor' instances, and the F1-score is 0.97.
- For class 2 ('virginica'), precision is 0.83, recall is 1.0, and the F1-score is 0.91.

- Overall, the model achieved an accuracy of 0.97, which indicates that it correctly predicted 97% of the instances in the test set.

```python
conf_matrix = confusion_matrix(y_test,predicted)
print("Confusion Matrix:\n", conf_matrix)
```

Total test data is 10 as 20% of 50 is 10

- The rows represent the actual class labels, and the columns represent the predicted class labels. The three classes are 'setosa', 'versicolor', and 'virginica', represented as 0, 1, and 2, respectively.

- Here's the interpretation for each row:

- Row 1: Actual class 'setosa' (class 0)

    True Positives (TP): 10

    True Negatives (TN): Not shown (TN for class 0 is the sum of the other elements in row 1)

- Row 2: Actual class 'versicolor' (class 1)

- Row 3: Actual class 'virginica' (class 2)

- Interpretation:

  - For class 'setosa' (class 0): The model correctly predicted all instances of 'setosa' (True Positives = 10). There are no False Positives or False Negatives for this class, as indicated by zeros in both those cells.
